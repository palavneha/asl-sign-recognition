```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model = load_model(r"C:\Users\Dell\Desktop\asl-sign-recognition\models\asl_model_v2.h5")
labels = np.load("models/labels.npy", allow_pickle=True)

print("Model loaded")
print("Classes:", labels)
```

```python
clean_labels = [label.replace("-samples", "") for label in labels]
clean_labels = np.array(clean_labels)

print("Clean labels:", clean_labels)
```

```python
def predict_image(img_path):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (64, 64))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)

    pred = model.predict(img)
    idx = np.argmax(pred)
    conf = np.max(pred)

    return clean_labels[idx], conf
```

```python
sentence = []

images = [

    r"C:\Users\Dell\Desktop\image.jpeg",    
    r"C:\Users\Dell\Desktop\image 1.jpeg",
    r"C:\Users\Dell\Desktop\image 2.jpeg"
]

for img in images:
    word, conf = predict_image(img)
    sentence.append(word)
    print(f"{word} ({conf*100:.2f}%)")

print("Final sentence:", " ".join(sentence))
```
